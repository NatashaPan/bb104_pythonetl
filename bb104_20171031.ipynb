```python
import requests

URL = "http://www.appledaily.com.tw/realtimenews/section/new/%s"

for page in range(1,11):
    tmp_url = URL%page
    resp = requests.get(tmp_url)
    print("[INFO] Crawling page %s, status is %s"%(tmp_url,resp.status_code))
    with open('./appledaily_%s.html'%page, 'w') as f:
        f.write(resp.text)
```

```python
# List or Dictionary comprehension
import random
scores = []
for x in range(10):
    scores.append(random.randint(1,10))
```

```python
smaller = []
for x in scores:
    if x < 3:
        x = x * 2
        smaller.append(x)
smaller
```

```python
smaller2 = [x*2 for x in scores if x < 3]
smaller2
```

```python
smaller2 = [x*2 for x in scores]
smaller2
```
